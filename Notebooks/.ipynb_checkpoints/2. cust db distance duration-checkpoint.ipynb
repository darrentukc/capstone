```python
import requests
import folium
import polyline
import pandas as pd
import time
```

```python
# import lat lng df

df = pd.read_csv('../data/df_lat_lng.csv')
```

```python
# function that returns route, distance and duration between 2 locations

def get_route(pickup_lon, pickup_lat, dropoff_lon, dropoff_lat):
    
    loc = "{},{};{},{}".format(pickup_lon, pickup_lat, dropoff_lon, dropoff_lat)
    url = "http://localhost:5000/route/v1/driving/"
    r = requests.get(url + loc) 
    if r.status_code!= 200:
        return {}
  
    res = r.json()   
    routes = polyline.decode(res['routes'][0]['geometry'])
    start_point = [res['waypoints'][0]['location'][1], res['waypoints'][0]['location'][0]]
    end_point = [res['waypoints'][1]['location'][1], res['waypoints'][1]['location'][0]]
    distance = round((res['routes'][0]['distance']/1000), 2)
    duration = round(res['routes'][0]['duration']/60)
    
    out = {'route':routes,
           'start_point':start_point,
           'end_point':end_point,
           'distance':distance,
           'duration':duration
          }

    return out
```

```python
# creating df for distance array

distance_array = pd.DataFrame(index = df['Name'].tolist(),
                              columns = df['Name'].tolist())
```

```python
# function to create distance matrix

def distance_matrix_generator(x):
    for i in range(len(x)):
        try:
            for j in range(len(x)):
                pickup_lon, pickup_lat, dropoff_lon, dropoff_lat = df.loc[i, 'Lng'], df.loc[i, 'Lat'], df.loc[j, 'Lng'], df.loc[j, 'Lat']
                current_route = get_route(pickup_lon, pickup_lat, dropoff_lon, dropoff_lat)
                row = x['Name'][i]
                column = x['Name'][j]
                distance_array.loc[row, column] = current_route['distance']
                print(i,j)
                time.sleep(0.05)
        except:
            pass
```

```python
distance_matrix_generator(df)
```

```python
# save distance matrix into csv for future use

distance_array.to_csv('../data/distance_array.csv')
```
